```python
import pandas as pd
from datetime import datetime as dt
import calendar
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
%matplotlib inline

import seaborn as sns
import pickle
```

```python
links=[]
dates=['190323','190330','190406','190413','190420','190427','190504','190511','190518','190525','190601']
links=[f"http://web.mta.info/developers/data/nyct/turnstile/turnstile_{date}.txt" for date in dates]
print (links[0])
print(len(links))
```

```python
df0=pd.read_csv(links[0])
df1=pd.read_csv(links[1])
df2=pd.read_csv(links[2])
df3=pd.read_csv(links[3])
df4=pd.read_csv(links[4])
df5=pd.read_csv(links[5])
df6=pd.read_csv(links[6])
df7=pd.read_csv(links[7])
df8=pd.read_csv(links[8])
df9=pd.read_csv(links[9])
df10=pd.read_csv(links[10])

mta_df=pd.concat([df0,df1,df2,df3,df4,df5,df6,df7,df8,df9,df10])
mta_df.head(5)

```

```python
mask=(mta_df["DESC"]== "REGULAR")
mta_df=mta_df[mask]
mta_df.head(10)
```

```python
from scipy import stats
df=pd.DataFrame(mta_df)
df['z_entry']=stats.zscore(df['ENTRIES'])
df.head(5)
df.loc[df['z_entry'].abs()<=3]
```

```python
mta_df.info()
```

```python
mta_df["DATETIME"]=pd.to_datetime(mta_df["DATE"]+" "+mta_df["TIME"],infer_datetime_format=True)
```

```python
#making columns consistent without spacing
mta_df.rename(columns={"C/A":"CA"}, inplace=True)
mta_df.columns=[column.strip() for column in mta_df.columns]
```

```python
#checking for completely duplicated rows
dupes=mta_df[mta_df.duplicated()]
```

```python
mta_df['LINENAME']=mta_df["LINENAME"].apply(lambda x:''.join(sorted(x)))
```

```python
#combining stations that are the same but were written differently
mta_df.loc[(mta_df.STATION=='1 AVE')& (mta_df.LINENAME=="L"),"STATION"]='1 AV'
mta_df.loc[(mta_df.STATION=='42 ST-GRD CNTRL')& (mta_df.LINENAME=='4567S'), "STATION"]="GRD CNTRL-42 ST"
```

```python
#creating unique turnstile column
mta_df["TURNSTILE"]=mta_df["CA"]+" "+mta_df["UNIT"]+" "+mta_df["SCP"]+" "+mta_df["STATION"]
mta_df["TURNSTILE"].head()
```

```python
##checking for turnstiles with duplicate entries for same datetime    
mta_df.groupby(['TURNSTILE','DATETIME']).ENTRIES.count().reset_index().sort_values("ENTRIES",ascending=False)
```

```python
#dropping those duplicates since dataframe is so large
mta_df.drop_duplicates(subset=["TURNSTILE","DATETIME"],inplace=True)
```

```python
#sanity check that all turnstile entries are individual
mta_df.groupby(['TURNSTILE','DATETIME']).ENTRIES.count().reset_index().sort_values("ENTRIES",ascending=False)
```

```python
#looking at first entries for each turnstiles
mta_daily=(mta_df
          .groupby(['CA','UNIT','SCP','STATION','TURNSTILE','DATE'],as_index=False)
          .ENTRIES.first())
mta_daily.head()
```

```python
mta_daily[["PREV_DATE","PREV_ENTRIES"]]=(mta_daily
                                        .groupby(["CA",'UNIT','SCP','STATION',"TURNSTILE"])['DATE','ENTRIES']
                                        .apply(lambda grp:grp.shift(1)))
```

```python
#removing NaN values
mta_daily.dropna(subset=["PREV_DATE"],axis=0,inplace=True)
```

```python
#compares previous current day to previous day's entries
mta_daily[mta_daily["ENTRIES"]<mta_daily["PREV_ENTRIES"]].head()
```

```python
(mta_daily[mta_daily["ENTRIES"]<mta_daily["PREV_ENTRIES"]].groupby(["CA","UNIT","SCP","STATION","TURNSTILE"]).size())
```

```python
def daily_counts(row,max_counter):
    '''
    Function to calculate daily entires, needed because entries were a cumulative value,
    max counter used to avoid outliers like unusually large entry values
    '''
    counter=row["ENTRIES"]-row["PREV_ENTRIES"]
    if counter <0:
        counter =-counter
    if counter>max_counter:
        print(row["ENTRIES"],row["PREV_ENTRIES"])
        return 0
    return counter
mta_daily["DAILY_ENTRIES"]=mta_daily.apply(daily_counts, axis=1, max_counter=100000)
```

```python
busiest_station=mta_daily.groupby(['STATION']).agg({'ENTRIES':'sum'}).reset_index()
busiest_station = busiest_station.sort_values(by='ENTRIES',ascending=False)
busiest_station.head(10)
```

```python
top_10_station_overall= (mta_daily.groupby(['STATION'])['DAILY_ENTRIES'].sum().reset_index().sort_values(by='DAILY_ENTRIES', ascending=False))
top_10_station_overall.head(10)
```

```python
# create MTA data #
final_mta_v1 = top_10_station_overall
```

```python
# normalizing 'DAILY_ENTRIES' #
final_mta_v1['NM_DAILY_ENTRIES']=(final_mta_v1['DAILY_ENTRIES']-final_mta_v1['DAILY_ENTRIES'].min())/(final_mta_v1['DAILY_ENTRIES'].max()-final_mta_v1['DAILY_ENTRIES'].min())
```
