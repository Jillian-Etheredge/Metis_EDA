```python
import pandas as pd
from datetime import datetime as dt
import calendar
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
%matplotlib inline

import seaborn as sns
import pickle
```

```python
links=[]
dates=['190323','190330','190406','190413','190420','190427','190504','190511','190518','190525','190601']
links=[f"http://web.mta.info/developers/data/nyct/turnstile/turnstile_{date}.txt" for date in dates]
print (links[0])
print(len(links))
```

```python
#loading data txt links into individual csv dataframes and concatinating into one main dataframe
df0=pd.read_csv(links[0])
df1=pd.read_csv(links[1])
df2=pd.read_csv(links[2])
df3=pd.read_csv(links[3])
df4=pd.read_csv(links[4])
df5=pd.read_csv(links[5])
df6=pd.read_csv(links[6])
df7=pd.read_csv(links[7])
df8=pd.read_csv(links[8])
df9=pd.read_csv(links[9])
df10=pd.read_csv(links[10])

mta_df=pd.concat([df0,df1,df2,df3,df4,df5,df6,df7,df8,df9,df10])
```

```python
#creating datetime column by converting date and time columns into datetime type
mta_df["DATETIME"]=pd.to_datetime(mta_df["DATE"]+" "+mta_df["TIME"],infer_datetime_format=True)
```

```python
mta_df.info()
```

```python
#making columns consistent without spacing
mta_df.rename(columns={"C/A":"CA"}, inplace=True)
mta_df.columns=[column.strip() for column in mta_df.columns]
mta_df.columns
```

```python
mta_df.head(10)
```

```python
#checking for completely duplicated rows
dupes=mta_df[mta_df.duplicated()]
dupes
```

```python
mta_df['LINENAME']=mta_df["LINENAME"].apply(lambda x:''.join(sorted(x)))
```

```python
mta_df.STATION.unique()
```

```python
#combining stations that are the same but were written differently
mta_df.loc[(mta_df.STATION=='1 AVE')& (mta_df.LINENAME=="L"),"STATION"]='1 AV'
mta_df.loc[(mta_df.STATION=='42 ST-GRD CNTRL')& (mta_df.LINENAME=='4567S'), "STATION"]="GRD CNTRL-42 ST"
```

```python
#creating unique turnstile column
mta_df["TURNSTILE"]=mta_df["CA"]+" "+mta_df["UNIT"]+" "+mta_df["SCP"]+" "+mta_df["STATION"]
mta_df["TURNSTILE"].head()
```

```python
##checking for turnstiles with duplicate entries for same datetime    
mta_df.groupby(['TURNSTILE','DATETIME']).ENTRIES.count().reset_index().sort_values("ENTRIES",ascending=False)
```

```python
#dropping those duplicates since dataframe is so large
mta_df.drop_duplicates(subset=["TURNSTILE","DATETIME"],inplace=True)
```

```python
#sanity check that all turnstile entries are individual
mta_df.groupby(['TURNSTILE','DATETIME']).ENTRIES.count().reset_index().sort_values("ENTRIES",ascending=False)
```

```python
#looking at first entries for each turnstiles
mta_daily=(mta_df
          .groupby(['CA','UNIT','SCP','STATION','TURNSTILE','DATE'],as_index=False)
          .ENTRIES.first())
mta_daily.head()
```

```python
mta_daily[["PREV_DATE","PREV_ENTRIES"]]=(mta_daily
                                        .groupby(["CA",'UNIT','SCP','STATION',"TURNSTILE"])['DATE','ENTRIES']
                                        .apply(lambda grp:grp.shift(1)))
mta_daily.head()
```

```python
#removing NaN values
mta_daily.dropna(subset=["PREV_DATE"],axis=0,inplace=True)
```

```python
#compares previous current day to previous day's entries
mta_daily[mta_daily["ENTRIES"]<mta_daily["PREV_ENTRIES"]].head()
```

```python
(mta_daily[mta_daily["ENTRIES"]<mta_daily["PREV_ENTRIES"]].groupby(["CA","UNIT","SCP","STATION","TURNSTILE"]).size())
```

```python
def daily_counts(row,max_counter):
    '''
    Function to calculate daily entires, needed because entries were a cumulative value,
    max counter used to avoid outliers like unusually large entry values
    '''
    counter=row["ENTRIES"]-row["PREV_ENTRIES"]
    if counter <0:
        counter =-counter
    if counter>max_counter:
        print(row["ENTRIES"],row["PREV_ENTRIES"])
        return 0
    return counter
mta_daily["DAILY_ENTRIES"]=mta_daily.apply(daily_counts, axis=1, max_counter=100000)
```

```python
## Challenge 5: ##
## for each ControlArea/UNIT/STATION combo, for each day, sum the counts from each turnstile belonging to that combo ##

mta_bycombo = mta_daily.groupby(['CA','UNIT','STATION','DATE']).agg({'DAILY_ENTRIES':'sum'}).reset_index()
mta_bycombo
```

```python
## Challenge 6: ##
## combine everything in each station, and come up with daily time series for each STATION, by adding up all the turnstiles in a station ##

mta_bystation=mta_bycombo.groupby(['STATION','DATE']).agg({'DAILY_ENTRIES':'sum'}).reset_index()
mta_bystation['STATION'].unique()
```

```python
mta_bystation=mta_bycombo.groupby(['STATION','DATE']).agg({'DAILY_ENTRIES':'sum'}).reset_index()
```

```python
## Challenge 7: ##
## Plot the daily time series for a station ##
##used every other date and daily entry as data became difficult to read with all dates included
##this had no noticable effect on overall graph shape
Station1=mta_bystation[mta_bystation['STATION']=='GRD CNTRL-42 ST']

fig=plt.figure(figsize=[10,10])
plt.plot(Station1['DATE'][::2],Station1['DAILY_ENTRIES'][::2]) 
plt.xlabel('Date', fontsize=15)
plt.xticks(fontsize=8)
plt.ylabel('Number of Entries',fontsize=15)
plt.title('Number of Entries for GRD CNTRL-42 ST',fontsize=20)
fig.autofmt_xdate()
plt.show()
plt.savefig('GRD_CNTRL_DAILY.svg',format='svg')
```

```python
## Challenge 8 ##
## Make one list of counts for one week for one station. Monday's count, Tuesday's count, etc. so it's a list of 7 counts. Make the same list for another week, and another week, and another week. ##
##Noticed that the original 'Weekday' column was incorrect, manually replaced values with correct ones
##May be a way to do this replacement with loop in future?
mta_daily['DATE']=pd.to_datetime(mta_daily['DATE'])
mta_daily['WEEKDAY'] =mta_daily['DATE'].dt.day_name()
mta_daily['WEEKDAY_INDEX'] = mta_daily['DATE'].dt.weekday
mta_daily['WEEK_OF_YEAR']=mta_daily['DATE'].dt.isocalendar().week
mta_daily['WEEKDAY'].replace({"Sunday":"Saturday1"}, inplace=True)
mta_daily['WEEKDAY'].replace({"Monday":"Sunday"}, inplace=True)
mta_daily['WEEKDAY'].replace({"Tuesday":"Monday"}, inplace=True)
mta_daily['WEEKDAY'].replace({"Wednesday":"Tuesday"}, inplace=True)
mta_daily['WEEKDAY'].replace({"Thursday":"Wednesday"}, inplace=True)
mta_daily['WEEKDAY'].replace({"Friday":"Thursday"}, inplace=True)
mta_daily['WEEKDAY'].replace({"Saturday":"Friday"}, inplace=True)
mta_daily['WEEKDAY'].replace({"Saturday1":"Saturday"}, inplace=True)
mta_daily.head(8)
```
