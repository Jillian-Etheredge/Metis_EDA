```python
import pandas as pd
import numpy as np
import pickle
```

```python
with open('df_education.pkl', 'rb') as read_file:
    df_education = pickle.load(read_file)
```

```python
df_education.set_index('GEO_ID')
```

```python
with open('df_gender_age.pkl', 'rb') as read_file:
    df_gender_age = pickle.load(read_file)
```

```python
df_gender_age.set_index('GEO_ID')
```

```python
df_census = df_education.join(df_gender_age, how = "outer", rsuffix = 'r')
df_census
```

```python
del df_census['GEO_IDr']
del df_census['NAMEr']
df_census
```

```python
#fill nulls with zero
df_census.fillna(value = 0, inplace = True)
```

```python
#create calculated field "Female Population Ratio"
df_census["Female Population Ratio"]=100/(df_census["Total population - Sex ratio (males per 100 females)"]+100)
df_census.drop(columns=["Total population - Sex ratio (males per 100 females)"], inplace = True)
```

```python
#create calculated field "Total Female Population - 25 years and over"
df_census["Total Female Population - 25 years and over"]= df_census["Total - AGE BY EDUCATIONAL ATTAINMENT - Population 25 years and over"]*df_census["Female Population Ratio"]
```

```python
#create calculated field "Total Female Population - 25 years and over - w/ advanced degree"
df_census["Total Female Population - 25 years and over - w/ advanced degree"]= df_census["Total - AGE BY EDUCATIONAL ATTAINMENT - Population 25 years and over - Bachelor's degree or higher"]*df_census["Female Population Ratio"]
```

```python
#rename desired columns
df_census.rename(columns={"Total - AGE BY EDUCATIONAL ATTAINMENT - Population 25 years and over":"Total Population - 25 years and over",
                          "Total - AGE BY EDUCATIONAL ATTAINMENT - Population 25 years and over - Bachelor's degree or higher":"Total Population - 25 years and over - Advanced Degrees",
                          "Total - MEDIAN EARNINGS IN THE PAST 12 MONTHS (IN 2019 INFLATION-ADJUSTED DOLLARS) - Population 25 years and over with earnings":"Median Income - 25 years and over"},
                inplace = True)
```

```python
df_census
```

```python
#prep GEO_ID field for later merge with NYC Open Data
df_census["GEO_ID"] = df_census["GEO_ID"].str[9:]
```
